```python
## Filtering df
```

```python
# Filtering is about rows
```

```python
stats.head()
```

```python
stats['Internet users'] < 2 # gives us all internet users where the value is smaller than 2
```

```python
Filter = stats['Internet users'] < 2
```

```python
stats[Filter] # use Filter on the stats object, this is the same as in R!!!
```

```python
#Awesome!
stats['Birth rate'] > 40
```

```python
Filter2 = stats['Birth rate'] > 40 #more than 40 children are born per every 1000 children per year
```

```python
stats[Filter2]
```

```python
# combine several filters
stats[stats['Birth rate'] > 40 and stats['Internet users'] < 2] # this is how you do it exactly in R
```

```python
stats['Birth rate'] > 40 and stats['Internet users'] < 2 ## Error is here!!!
```

```python
stats[Filter & Filter2] # use the bitwise & to make these operations!
```

```python
stats[(stats['Birth rate'] > 40) & (stats['Internet users'] < 2)] 
```

```python
# Another one
stats[(stats['Income Group'] == 'High Income')] ## for boolean operators the '==' has to be taken
```
